```python
# Import libraries
import pandas as pd
import pandasql as ps
from pandasql import sqldf
pysqldf = lambda q: sqldf(q, globals())
```

```python
df = pd.read_csv('/content/drive/MyDrive/all_csv sorted.csv')
```

```python
df.columns
```

# Preparing Data

```python
# Drop columns
df = df.drop(['Unnamed: 0'], axis=1)
df = df.drop(['Country code'], axis=1)
```

```python
# Rename columns
df.rename(columns = {'Continental region':'Continental_region', 
                     'NO. OF Internet Plans ':'NO._OF_Internet_Plans',
                     'Average price of 1GB (USD)':'Average_price_of_1GB_USD',
                     'Cheapest 1GB for 30 days (USD)':'Cheapest_1GB_for_30_days_USD', 
                     'Most expensive 1GB (USD)':'Most_expensive_1GB_USD',
                     'Average price of 1GB (USD  at the start of 2021)':'Average_price_of_1GB_USD_at_the_start_of_2021',
                     'Average price of 1GB (USD – at start of 2020)':'Average_price_of_1GB_USD_at_start_of_2020',
                     'Internet users':'Internet_Users',
                     'Avg \n(Mbit/s)Ookla':'Speed_in_Mb/s'}, inplace = True)
```

```python
df.tail()
```

```python
# Check the null values
df.info()
```

```python
# Drop null values
df.drop(df[df['Population'].isnull()].index,inplace = True)

df.drop(df[df['Average_price_of_1GB_USD'] == 'NO PROVIDERS'].index,inplace = True)
df.drop(df[df['Average_price_of_1GB_USD'] == 'HYPERINFLATION'].index,inplace = True)
df.drop(df[df['Average_price_of_1GB_USD'] == "Prices listed in non-convertible 'units'"].index,inplace = True)

df.drop(df[df['NO._OF_Internet_Plans'].isnull()].index,inplace = True)

df.drop(df[df['Average_price_of_1GB_USD_at_start_of_2020'] == 'NO PACKAGES'].index,inplace = True)
df.drop(df[df['Average_price_of_1GB_USD_at_the_start_of_2021'] == 'NO PACKAGES'].index,inplace = True)
df.drop(df[df['Speed_in_Mb/s'].isnull()].index,inplace = True)
```
