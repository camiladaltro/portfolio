### 1. Objective
Analyze data from a supermarket franchise to find out how sales are doing and how it relates to other aspects.

### 2. Import and Load the Data

```python
# Import libraries for data analysis
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the data
df = pd.read_csv('/content/drive/MyDrive/Stores.csv')
```

```python
# Check the info
df.info()
```

```python
# Overview
df.head()
```

### 3. Exploratory Data Analysis
* General descriptive statistics

```python
# general descriptive statistics
df.describe().T
```

* How do the columns relate to each other?

```python
# Correlation matrix
sns.heatmap(df.corr(),annot=True,cmap='crest',linewidths=0.2)
fig=plt.gcf()
fig.set_size_inches(10,6)
plt.show()
```

```python
# area x items plot visualization
plt.figure(figsize=(9,6))
sns.scatterplot(data=df, x="Store_Area", y="Items_Available", color='mediumseagreen')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(9,6))
sns.scatterplot(data=df, x="Store_Area", y="Store_Sales", color='seagreen')
plt.xticks(rotation=90)
plt.show()
```

```python
# items x sales plot visualization
plt.figure(figsize=(9,6))
sns.scatterplot(data=df, x="Items_Available", y="Store_Sales", color='lightseagreen')
plt.xticks(rotation=90)
plt.show()
```

```python
# customers x sales plot visualization
plt.figure(figsize=(9,6))
sns.scatterplot(data=df, x="Daily_Customer_Count", y="Store_Sales", color='darkblue')
plt.xticks(rotation=90)
plt.show()
```

* How many stores sell above average?

```python
# check the number of stores that sell above average
df[df.Store_Sales > 59351].sort_values(by='Store_Sales', ascending=False).info()
```

```python
# check the number of stores that sell below average
df[df.Store_Sales < 59351].sort_values(by='Store_Sales', ascending=False).info()
```

### 5. Conclusions
* The stores area is perfectly related to the availability of items, while the other aspects have a low correlation. If the store has a large area, the availability of items is also.
* The other variables are not dependent (eg. number of customers doesn't influence sales, available items don't affect sales, etc.)
* Most stores are selling below average (463 stores).
